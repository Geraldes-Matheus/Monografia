## geração de jatos usando o fastjet em um arquivo hepmc

```python
import pyhepmc as hep
import numpy as np
import hist
import matplotlib.pyplot as plt
import fastjet as fj

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Definir histogramas
histo_njets  = hist.Hist.new.Reg(10, 0., 20.).Double()
histo_jets_pt  = hist.Hist.new.Reg(50, 0., 200.).Double()
histo_jets_eta = hist.Hist.new.Reg(500, -5., +5.).Double()
histo_jets_mass = hist.Hist.new.Reg(50, 0., 200.).Double()  # Massa invariante dos jatos
histo_npart  = hist.Hist.new.Reg(50, 0., 5000.).Double()
histo_jets_energy = hist.Hist.new.Reg(50, 0., 500.).Double()  # Energia dos jatos
histo_rjj = hist.Hist.new.Reg(50, 0., 5.0).Double()  # Separação entre jatos (Rjj)

# Função para ler o arquivo HEPMC e extrair os eventos
def read_hepmc_file(hepmc_file):
    events = []
    with hep.open(hepmc_file) as f:
        for event in f:
            events.append(event)  # Adiciona o evento ao arquivo
    return events

# Função para calcular o momento transversal (pT)
def get_pt(particle):
    return np.sqrt(particle.momentum.x**2 + particle.momentum.y**2)

# Função para calcular a pseudorapidez (eta)
def get_eta(particle):
    p = np.sqrt(particle.momentum.x**2 + particle.momentum.y**2 + particle.momentum.z**2)
    if p == 0 or np.isclose(particle.momentum.z, p) or np.isclose(particle.momentum.z, -p):
        return np.nan  # Retorna NaN se não puder calcular
    return 0.5 * np.log((p + particle.momentum.z) / (p - particle.momentum.z))

# Função para calcular a energia de uma partícula
def get_energy(particle):
    return particle.momentum.e

# Função para calcular a separação Rjj entre dois jatos
def delta_r(jet1, jet2):
    delta_eta = jet1.eta() - jet2.eta()
    delta_phi = jet1.phi() - jet2.phi()
    return np.sqrt(delta_eta**2 + delta_phi**2)

# Função para calcular a massa invariante de dois jatos
def invariant_mass(jet1, jet2):
    E_tot = jet1.e() + jet2.e()
    px_tot = jet1.px() + jet2.px()
    py_tot = jet1.py() + jet2.py()
    pz_tot = jet1.pz() + jet2.pz()
    return np.sqrt(E_tot**2 - (px_tot**2 + py_tot**2 + pz_tot**2))

# Função para analisar o evento e preencher os histogramas de jatos
def analyze_event(event):
    # Filtrar partículas com pT > 10 GeV para reconstruir jatos
    particles = [p for p in event.particles if p.status == 1]  # Somente partículas finais (status 1)
    
    # Converter partículas para objetos PseudoJet do FastJet
    fj_particles = []
    for p in particles:
        fj_particle = fj.PseudoJet(p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e)
        fj_particles.append(fj_particle)

    # Escolher o algoritmo de clustering (anti-kT, por exemplo)
    jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)  # Radius de 0.4 
    
    # Clustering dos jatos
    clusterer = fj.ClusterSequence(fj_particles, jet_def)
    jets = clusterer.inclusive_jets(10.0)  # Jatos com pT > 10 GeV 
    # Preencher histogramas de jatos
    njets = len(jets)
    histo_njets.fill(njets)
    
    # Preencher histogramas de pT, eta, energia e massa dos jatos
    for jet in jets:
        histo_jets_pt.fill(jet.pt())  # Preencher o histograma de pT dos jatos
        histo_jets_eta.fill(jet.eta())  # Preencher o histograma de eta dos jatos
        histo_jets_energy.fill(jet.e())  # Preencher o histograma de energia dos jatos
        
    # Preencher histograma de massa invariante dos jatos
    if len(jets) >= 2:
        for i in range(len(jets)-1):
            for j in range(i+1, len(jets)):
                mass = invariant_mass(jets[i], jets[j])
                histo_jets_mass.fill(mass)

                # Preencher histograma de separação Rjj entre os jatos
                rjj = delta_r(jets[i], jets[j])
                histo_rjj.fill(rjj)

    # Número de partículas no evento
    npart = len(particles)
    histo_npart.fill(npart)

# Função para visualizar os histogramas
def plot_histograms():
    fig, ax = plt.subplots(3, 2, figsize=(12, 14))

    histo_npart.plot(ax=ax[0, 0], ls="-", color="blue", lw=3)
    ax[0, 0].set(xlabel='Número de Partículas', ylabel='Contagem')

    histo_njets.plot(ax=ax[0, 1], ls="-", color="blue", lw=3)
    ax[0, 1].set(xlabel='Número de Jatos', ylabel='Contagem')

    histo_jets_pt.plot(ax=ax[1, 0], ls="-", color="red", lw=3)
    ax[1, 0].set(xlabel='pT dos Jatos (GeV)', ylabel='Contagem')

    histo_jets_eta.plot(ax=ax[1, 1], ls="-", color="red", lw=3)
    ax[1, 1].set(xlabel='Rapidez dos Jatos (Eta)', ylabel='Contagem')

    histo_jets_mass.plot(ax=ax[2, 0], ls="-", color="green", lw=3)
    ax[2, 0].set(xlabel='Massa Invariante dos Jatos (GeV)', ylabel='Contagem')

    # Plot para a Distribuição de Energia dos Jatos no eixo [2,1]
    histo_jets_energy.plot(ax=ax[2, 1], ls="-", color="purple", lw=3, density=True)
    ax[2, 1].set(xlabel='Energia dos Jatos (GeV)', ylabel='Probabilidade')

    # Histograma de separação Rjj entre jatos
    plt.figure(figsize=(8, 6))
    histo_rjj.plot(ls="-", color="purple", lw=3)
    plt.xlabel('Separação entre Jatos Rjj')
    plt.ylabel('Número de Pares de Jatos')
    plt.title('Distribuição de Separação entre Jatos (Rjj)')
    plt.show()

# Ler eventos do arquivo HEPMC
events = read_hepmc_file(hepmc_file)

# Loop sobre os eventos e realizar a análise
nevt = len(events)
for ievt in range(nevt):
    if ievt % 1000 == 0:
        print(f"Processando evento {ievt} de {nevt}...")

    event = events[ievt]

    # Analisar o evento e preencher os histogramas
    analyze_event(event)

# Plotar os histogramas
plot_histograms()
```

```python
import pyhepmc as hep
import numpy as np
import hist
import matplotlib.pyplot as plt
import fastjet as fj

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Definir histogramas
# ... [Outros histogramas]

# Função para ler o arquivo HEPMC e extrair os eventos
def read_hepmc_file(hepmc_file):
    events = []
    with hep.open(hepmc_file) as f:
        for event in f:
            events.append(event)  # Adiciona o evento ao arquivo
    print(f"Leitura de {len(events)} eventos do arquivo HEPMC.")
    return events

# Função para obter o vértice primário (assumimos o primeiro vértice)
def get_primary_vertex(event):
    for vertex in event.vertices:
        # O vértice primário é geralmente o primeiro
        return vertex.position
    return None  # Caso não haja vértices, retornar None

# Função para analisar o evento e preencher os histogramas de jatos
def analyze_event(event):
    # Obter o vértice primário
    primary_vertex = get_primary_vertex(event)
    if primary_vertex is None:
        return [], []  # Se não houver vértice primário, pular o evento

    
    particles = [p for p in event.particles if p.status == 1]  # Somente partículas finais (status 1)
    
    # Converter partículas para objetos PseudoJet do FastJet
    fj_particles = []
    for p in particles:
        fj_particle = fj.PseudoJet(p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e)
        fj_particles.append(fj_particle)

    # Escolher o algoritmo de clustering (anti-kT, por exemplo)
    jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)  # Radius de 0.4 
    
    # Clustering dos jatos
    clusterer = fj.ClusterSequence(fj_particles, jet_def)
    jets = clusterer.inclusive_jets(10.0)  # Jatos com pT > 10 GeV 
    
    # Preencher histogramas de jatos
    # ... [Preenchimento dos histogramas]

    # Calcular a posição dos jatos em relação ao vértice primário
    jet_positions_x = []
    jet_positions_y = []
    for jet in jets:
        # Posição do jato (x, y) em relação ao vértice primário
        # Para isso, calculamos a posição 2D no plano (transversal) usando pT e phi
        pt = jet.pt()  # pT do jato
        phi = jet.phi()  # ângulo azimutal do jato

        # Calculando as coordenadas (x, y) do jato
        jet_x = pt * np.cos(phi) - primary_vertex.x  
        jet_y = pt * np.sin(phi) - primary_vertex.y  
        
        jet_positions_x.append(jet_x)
        jet_positions_y.append(jet_y)

    return jet_positions_x, jet_positions_y

# Função para visualizar os histogramas e plotar as posições dos jatos
def plot_histograms(jet_positions_x, jet_positions_y):
    # Subplots dos histogramas de jatos
    fig, ax = plt.subplots(3, 2, figsize=(12, 14))

    histo_npart.plot(ax=ax[0, 0], ls="-", color="blue", lw=3)
    ax[0, 0].set(xlabel='Número de Partículas', ylabel='Contagem')

    histo_njets.plot(ax=ax[0, 1], ls="-", color="blue", lw=3)
    ax[0, 1].set(xlabel='Número de Jatos', ylabel='Contagem')

    histo_jets_pt.plot(ax=ax[1, 0], ls="-", color="red", lw=3)
    ax[1, 0].set(xlabel='pT dos Jatos (GeV)', ylabel='Contagem')

    histo_jets_eta.plot(ax=ax[1, 1], ls="-", color="red", lw=3)
    ax[1, 1].set(xlabel='Rapidez dos Jatos (Eta)', ylabel='Contagem')

    histo_jets_mass.plot(ax=ax[2, 0], ls="-", color="green", lw=3)
    ax[2, 0].set(xlabel='Massa Invariante dos Jatos (GeV)', ylabel='Contagem')

    histo_bjets_pt.plot(ax=ax[2, 1], ls="-", color="orange", lw=3)
    ax[2, 1].set(xlabel='pT dos b-Jatos (GeV)', ylabel='Contagem')

    # Histograma de massa dos b-jatos
    plt.figure(figsize=(8, 6))
    histo_bjets_mass.plot(ls="-", color="purple", lw=3)
    plt.xlabel('Massa Invariante dos b-Jatos (GeV)')
    plt.ylabel('Número de b-Jatos')
    plt.title('Distribuição de Massa dos b-Jatos')
    plt.show()

    # Plotar a posição dos jatos em relação ao vértice primário
    plt.figure(figsize=(8, 6))
    plt.scatter(jet_positions_x, jet_positions_y, color='blue', alpha=0.5, s=10)
    plt.xlabel('Distância Radial do Jato (x) em relação ao vértice primário (cm)')
    plt.ylabel('Distância Radial do Jato (y) em relação ao vértice primário (cm)')
    plt.title('Posição dos Jatos em Relação ao Vértice Primário')
    plt.grid(True)
    plt.show()

# Ler eventos do arquivo HEPMC
events = read_hepmc_file(hepmc_file)

# Lista para armazenar as posições dos jatos
all_jet_positions_x = []
all_jet_positions_y = []

# Loop sobre os eventos e realizar a análise
nevt = len(events)
for ievt in range(nevt):
    if ievt % 1000 == 0:
        print(f"Processando evento {ievt + 1}/{nevt}...")
    jet_positions_x, jet_positions_y = analyze_event(events[ievt])
    all_jet_positions_x.extend(jet_positions_x)
    all_jet_positions_y.extend(jet_positions_y)

# Exibir os histogramas
plot_histograms(all_jet_positions_x, all_jet_positions_y)
```

```python
import pyhepmc as hep
import numpy as np
import hist
import matplotlib.pyplot as plt
import fastjet as fj

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Definir histogramas para análise do parâmetro de impacto
histo_impact_param = hist.Hist.new.Reg(50, 0., 0.05).Double()  # Parâmetro de impacto
histo_b_jets = hist.Hist.new.Reg(50, 0., 0.05).Double()  # Parâmetro de impacto para jatos b
histo_non_b_jets = hist.Hist.new.Reg(50, 0., 0.05).Double()  # Parâmetro de impacto para jatos não-b
histo_high_impact_count = hist.Hist.new.Reg(10, 0, 10).Double()  # Número de partículas com |d₀| > 0.01 cm

# Função para ler o arquivo HEPMC e extrair os eventos
def read_hepmc_file(hepmc_file):
    events = []
    with hep.open(hepmc_file) as f:
        for event in f:
            events.append(event)  # Adiciona o evento ao arquivo
    return events

# Função para simular o parâmetro de impacto de uma partícula (exemplo simples)
def simulate_impact_parameter(particle):
    # Aqui você pode adicionar um modelo mais complexo, dependendo dos dados do detector
    return np.abs(particle.momentum.x * particle.momentum.y) / np.sqrt(particle.momentum.x**2 + particle.momentum.y**2)

# Função para verificar se o jato é um jato de b-quark (exemplo simples)
def is_b_jet_by_impact(jet, particles):
    # Aqui você pode usar alguma estratégia para identificar jatos de b-quarks
    # Exemplo: verificar se o jato contém partículas com grande parâmetro de impacto
    for p in particles:
        impact = simulate_impact_parameter(p)
        if impact > 0.01:
            return True  # Se encontrar uma partícula com impacto significativo, consideramos o jato como b-jet
    return False

# Função para analisar o evento e preencher os histogramas de jatos
def analyze_event(event):
    # Filtrar partículas com pT > 10 GeV para reconstruir jatos
    particles = [p for p in event.particles if p.status == 1]  # Somente partículas finais (status 1)
    
    # Converter partículas para objetos PseudoJet do FastJet
    fj_particles = []
    for p in particles:
        fj_particle = fj.PseudoJet(p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e)
        fj_particles.append(fj_particle)

    # Escolher o algoritmo de clustering (anti-kT, por exemplo)
    jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)  # Radius de 0.4 
    
    # Clustering dos jatos
    clusterer = fj.ClusterSequence(fj_particles, jet_def)
    jets = clusterer.inclusive_jets(10.0)  # Jatos com pT > 10 GeV 
    # Preencher histogramas de jatos
    njets = len(jets)
    histo_njets.fill(njets)
    
    # Análise do parâmetro de impacto para o histograma
    for p in particles:
        impact = simulate_impact_parameter(p)
        histo_impact_param.fill(impact)
        
        # Análise de jatos b e não-b
        if is_b_jet_by_impact(jets[0], particles):  # Verifica se o primeiro jato é b-jet
            histo_b_jets.fill(impact)
        else:
            histo_non_b_jets.fill(impact)

    # Análise de partículas com alto parâmetro de impacto
    for jet in jets:
        associated_particles = [p for p in particles if delta_r(jet, fj.PseudoJet(p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e)) < 0.4]
        high_impact_count = sum(1 for p in associated_particles if simulate_impact_parameter(p) > 0.01)
        histo_high_impact_count.fill(high_impact_count)

# Função para visualizar os histogramas
def plot_histograms():
    fig, ax = plt.subplots(2, 2, figsize=(12, 12))

    histo_impact_param.plot(ax=ax[0, 0], ls="-", color="blue", lw=3)
    ax[0, 0].set(xlabel='Parâmetro de Impacto', ylabel='Contagem')

    histo_b_jets.plot(ax=ax[0, 1], ls="-", color="red", lw=3)
    ax[0, 1].set(xlabel='Parâmetro de Impacto para Jatos b', ylabel='Contagem')

    histo_non_b_jets.plot(ax=ax[1, 0], ls="-", color="green", lw=3)
    ax[1, 0].set(xlabel='Parâmetro de Impacto para Jatos não-b', ylabel='Contagem')

    histo_high_impact_count.plot(ax=ax[1, 1], ls="-", color="purple", lw=3)
    ax[1, 1].set(xlabel='Número de Partículas com Alto Impacto', ylabel='Contagem')

    plt.tight_layout()
    plt.show()

# Ler eventos do arquivo HEPMC
events = read_hepmc_file(hepmc_file)

# Loop sobre os eventos e realizar a análise
nevt = len(events)
for ievt in range(nevt):
    if ievt % 1000 == 0:
        print(f"Processando evento {ievt} de {nevt}...")

    event = events[ievt]

    # Analisar o evento e preencher os histogramas
    analyze_event(event)

# Plotar os histogramas
plot_histograms()
```

# Estratégia de Identificação

## Distância de Decaimento (Impact Parameter)

- Hadron $b$ possuem vida média longa, resultando em vértices secundários ou terciários deslocados.
- Trilha com grande parâmetro de impacto (IP) é indicativa de jatos $b$.

## Produtos de Decaimento

- Hadron $b$ frequentemente decaem semi-leptonicamente (múon ou elétron com alto $p_T$ e $\Delta R$ pequeno em relação ao jato).

## Distribuição de Massa e Multiplicidade

- Hadron $b$ têm massa alta (~5 GeV) e produzem múltiplas partículas carregadas.

```python
import pyhepmc as hep
import numpy as np
import hist
import matplotlib.pyplot as plt
import fastjet as fj

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Função para calcular o parâmetro de impacto
def calculate_impact_parameter(particle, primary_vertex):
    # Acesse as coordenadas do vértice de produção e do vértice principal
    displacement = np.array([
        particle.production_vertex.position.x - primary_vertex.position.x,
        particle.production_vertex.position.y - primary_vertex.position.y,
        particle.production_vertex.position.z - primary_vertex.position.z
    ])
    momentum = np.array([particle.momentum.x, particle.momentum.y, particle.momentum.z])
    # Calcule o parâmetro de impacto
    impact_param = np.linalg.norm(np.cross(displacement, momentum)) / np.linalg.norm(momentum)
    return impact_param

# Função para verificar presença de léptons nos jatos
def has_lepton_in_jet(jet, particles):
    for particle in particles:
        # Use particle.pid para acessar o PDG ID
        if abs(particle.pid) in [11, 13]:  # IDs de elétron e múon
            lepton_vector = np.array([particle.momentum.x, particle.momentum.y, particle.momentum.z])
            jet_vector = np.array([jet.px(), jet.py(), jet.pz()])
            delta_r = np.sqrt((jet.eta() - get_eta(particle))**2 + (jet.phi() - np.arctan2(particle.momentum.y, particle.momentum.x))**2)
            if delta_r < 0.4:  # Critério de isolamento do lépton no jato
                return True
    return False

# Função para analisar o evento e identificar jatos b
def analyze_event(event):
    primary_vertex = event.vertices[0]
    particles = [p for p in event.particles if p.status == 1]  # Somente partículas finais

    fj_particles = []
    for p in particles:
        fj_particle = fj.PseudoJet(p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e)
        fj_particle.set_user_index(p.id)  # Vincular partícula ao PseudoJet
        fj_particles.append(fj_particle)

    jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)
    clusterer = fj.ClusterSequence(fj_particles, jet_def)
    jets = clusterer.inclusive_jets(10.0)

    b_tagged_jets = []
    for jet in jets:
        # Identificar jatos b com base em IP médio e léptons associados
        jet_particles = [p for p in particles if fj.PseudoJet(p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e).delta_R(jet) < 0.4]
        avg_impact_param = np.mean([calculate_impact_parameter(p, primary_vertex) for p in jet_particles])
        has_lepton = has_lepton_in_jet(jet, particles)
        
        # Critério de b-tagging
        if avg_impact_param > 0.05 or has_lepton:
            b_tagged_jets.append(jet)

    return b_tagged_jets

# Ler eventos do arquivo HEPMC
events = read_hepmc_file(hepmc_file)

# Loop sobre eventos
for ievt, event in enumerate(events):
    if ievt % 1000 == 0:
        print(f"Processando evento {ievt}...")
    b_jets = analyze_event(event)
    print(f"Evento {ievt}: {len(b_jets)} jatos b identificados.")
```

```python
import pyhepmc as hep
import numpy as np
import hist
import matplotlib.pyplot as plt
import fastjet as fj

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Função para calcular o parâmetro de impacto
def calculate_impact_parameter(particle, primary_vertex):
    displacement = np.array([
        particle.production_vertex.position.x - primary_vertex.position.x,
        particle.production_vertex.position.y - primary_vertex.position.y,
        particle.production_vertex.position.z - primary_vertex.position.z
    ])
    momentum = np.array([particle.momentum.x, particle.momentum.y, particle.momentum.z])
    impact_param = np.linalg.norm(np.cross(displacement, momentum)) / np.linalg.norm(momentum)
    return impact_param

# Função para verificar presença de léptons nos jatos
def has_lepton_in_jet(jet, particles):
    for particle in particles:
        if abs(particle.pid) in [11, 13]:  # IDs de elétron e múon
            delta_r = np.sqrt(
                (jet.eta() - get_eta(particle))**2 +
                (jet.phi() - np.arctan2(particle.momentum.y, particle.momentum.x))**2
            )
            if delta_r < 0.4:  # Critério de isolamento do lépton no jato
                return True
    return False

# Função para calcular a pseudorapidez
def get_eta(particle):
    p = np.sqrt(particle.momentum.x**2 + particle.momentum.y**2 + particle.momentum.z**2)
    if p == 0 or np.isclose(particle.momentum.z, p) or np.isclose(particle.momentum.z, -p):
        return np.nan
    return 0.5 * np.log((p + particle.momentum.z) / (p - particle.momentum.z))

# Função para analisar o evento e identificar jatos b
def analyze_event(event):
    primary_vertex = event.vertices[0]
    particles = [p for p in event.particles if p.status == 1]

    fj_particles = []
    for p in particles:
        fj_particle = fj.PseudoJet(p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e)
        fj_particle.set_user_index(p.id)
        fj_particles.append(fj_particle)

    jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)
    clusterer = fj.ClusterSequence(fj_particles, jet_def)
    jets = clusterer.inclusive_jets(10.0)

    b_tagged_jets = []
    non_b_jets = []
    for jet in jets:
        jet_particles = [p for p in particles if fj.PseudoJet(
            p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e).delta_R(jet) < 0.4]
        avg_impact_param = np.mean([calculate_impact_parameter(p, primary_vertex) for p in jet_particles])
        has_lepton = has_lepton_in_jet(jet, particles)
        if avg_impact_param > 0.05 or has_lepton:
            b_tagged_jets.append(jet)
        else:
            non_b_jets.append(jet)
    return b_tagged_jets, non_b_jets

# Função para plotar os jatos
def plot_jets(b_jets, non_b_jets):
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Plotar jatos b-tagged
    b_eta = [jet.eta() for jet in b_jets]
    b_phi = [jet.phi() for jet in b_jets]
    ax.scatter(b_eta, b_phi, c='red', label='b-tagged Jets', alpha=0.7)

    # Plotar jatos não b-tagged
    non_b_eta = [jet.eta() for jet in non_b_jets]
    non_b_phi = [jet.phi() for jet in non_b_jets]
    ax.scatter(non_b_eta, non_b_phi, c='blue', label='Non b-tagged Jets', alpha=0.7)

    ax.set_xlabel('Pseudorapidez (η)')
    ax.set_ylabel('Ângulo Azimutal (φ)')
    ax.set_title('Distribuição de Jatos')
    ax.legend()
    plt.grid()
    plt.show()

    # Histogramas
    plt.figure(figsize=(12, 8))
    plt.hist([jet.pt() for jet in b_jets], bins=50, alpha=0.7, label='b-tagged Jets', color='red')
    plt.hist([jet.pt() for jet in non_b_jets], bins=50, alpha=0.7, label='Non b-tagged Jets', color='blue')
    plt.xlabel('pT (GeV)')
    plt.ylabel('Contagem')
    plt.title('Distribuição de pT dos Jatos')
    plt.legend()
    plt.grid()
    plt.show()

# Ler eventos do arquivo HEPMC
events = read_hepmc_file(hepmc_file)

# Loop sobre eventos
all_b_jets = []
all_non_b_jets = []
for ievt, event in enumerate(events):
    if ievt % 1000 == 0:
        print(f"Processando evento {ievt}...")
    b_jets, non_b_jets = analyze_event(event)
    all_b_jets.extend(b_jets)
    all_non_b_jets.extend(non_b_jets)

# Plotar resultados
plot_jets(all_b_jets, all_non_b_jets)
```

## Critério do pTrel

Múons provenientes do decaimento de b-hadrons tendem a ter valores de \( p_T \) transversais em relação ao jato significativamente maiores que múons de fundo.

## Delta R

O critério de distância \( \Delta R \) entre o jato e partículas associadas ajuda a identificar múons que pertencem ao jato.


```python
import pyhepmc as hep
import numpy as np
import hist
import matplotlib.pyplot as plt
import fastjet as fj

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Definir histogramas para b-jatos
histo_bjets_pt = hist.Hist.new.Reg(50, 0., 200.).Double()
histo_bjets_eta = hist.Hist.new.Reg(500, -5., +5.).Double()
histo_bjets_mass = hist.Hist.new.Reg(50, 0., 200.).Double()  # Massa invariante dos b-jatos

# Função para calcular o momento transversal (pT)
def get_pt(particle):
    return np.sqrt(particle.momentum.x**2 + particle.momentum.y**2)

# Função para calcular a pseudorapidez (eta)
def get_eta(particle):
    p = np.sqrt(particle.momentum.x**2 + particle.momentum.y**2 + particle.momentum.z**2)
    if p == 0 or np.isclose(particle.momentum.z, p) or np.isclose(particle.momentum.z, -p):
        return np.nan
    return 0.5 * np.log((p + particle.momentum.z) / (p - particle.momentum.z))

# Função para calcular a separação Rjj entre dois jatos
def delta_r(jet, particle):
    delta_eta = jet.eta() - get_eta(particle)
    delta_phi = jet.phi() - np.arctan2(particle.momentum.y, particle.momentum.x)
    delta_phi = (delta_phi + np.pi) % (2 * np.pi) - np.pi  # Ajuste para o intervalo [-π, π]
    return np.sqrt(delta_eta**2 + delta_phi**2)

# Função para calcular o pTrel
def pt_rel(particle, jet):
    p_particle = np.array([particle.momentum.x, particle.momentum.y, particle.momentum.z])
    p_jet = np.array([jet.px(), jet.py(), jet.pz()])
    p_paralelo = np.dot(p_particle, p_jet) / np.linalg.norm(p_jet)
    p_rel = np.sqrt(np.linalg.norm(p_particle)**2 - p_paralelo**2)
    return p_rel

# Função para verificar se uma partícula é um múon
def is_muon(particle):
    return abs(particle.pid) == 13  # Use 'pid' em vez de 'pdg_id'

# Função para identificar b-jatos com base em critérios semileptônicos
def identify_b_jets(jets, particles):
    b_jets = []
    for jet in jets:
        associated_muons = [
            p for p in particles if is_muon(p) and delta_r(jet, p) < 0.4
        ]
        if len(associated_muons) > 0:
            muon_pt_rel = [pt_rel(muon, jet) for muon in associated_muons]
            if max(muon_pt_rel) > 0.5:  # Critério de pTrel
                b_jets.append(jet)
    return b_jets

# Função para analisar o evento e preencher os histogramas de b-jatos
def analyze_event(event):
    # Filtrar partículas com pT > 10 GeV para reconstruir jatos
    particles = [p for p in event.particles if p.status == 1]
    
    # Converter partículas para objetos PseudoJet do FastJet
    fj_particles = [
        fj.PseudoJet(p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e) for p in particles
    ]

    # Escolher o algoritmo de clustering (anti-kT, por exemplo)
    jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)
    clusterer = fj.ClusterSequence(fj_particles, jet_def)
    jets = clusterer.inclusive_jets(10.0)  # Jatos com pT > 10 GeV 
    
    # Identificar b-jatos
    b_jets = identify_b_jets(jets, particles)
    
    # Preencher histogramas de b-jatos
    for b_jet in b_jets:
        histo_bjets_pt.fill(b_jet.pt())  # Preencher o histograma de pT dos b-jatos
        histo_bjets_eta.fill(b_jet.eta())  # Preencher o histograma de eta dos b-jatos

# Função para visualizar os histogramas de b-jatos
def plot_histograms():
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    histo_bjets_pt.plot(ax=ax[0], ls="-", color="red", lw=3)
    ax[0].set(xlabel='pT dos b-Jatos (GeV)', ylabel='Contagem')

    histo_bjets_eta.plot(ax=ax[1], ls="-", color="blue", lw=3)
    ax[1].set(xlabel='Rapidez dos b-Jatos (Eta)', ylabel='Contagem')

    plt.tight_layout()
    plt.show()

# Ler eventos do arquivo HEPMC
events = []
with hep.open(hepmc_file) as f:
    for event in f:
        events.append(event)

# Loop sobre os eventos e realizar a análise
for ievt, event in enumerate(events):
    if ievt % 1000 == 0:
        print(f"Processando evento {ievt} de {len(events)}...")
    analyze_event(event)

# Plotar os histogramas de b-jatos
plot_histograms()
```

## 1.Deslocamento do Vértice Secundário e Parâmetros de Impacto

  Calcular o deslocamento das trilhas em relação ao vértice primário.
  Determinar a significância do parâmetro de impacto:
    A significância é dada pela razão entre o deslocamento e sua incerteza.

## 2.Massa do Vértice Secundário

 Reconstruir a massa invariante a partir das trilhas associadas ao vértice secundário.

## 3.Decaimentos Semileptônicos

 Identificar os múons e elétrons associados aos jatos.

## 4.Distribuições Cintemáticas

 Focar nos jatos com maior pT e rápida fragmentação.

```python
import pyhepmc as hep
import numpy as np
import hist
import matplotlib.pyplot as plt
import fastjet as fj

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Histogramas para b-jatos
histo_bjets_pt = hist.Hist.new.Reg(50, 0., 200.).Double()
histo_bjets_eta = hist.Hist.new.Reg(500, -5., +5.).Double()
histo_bjets_mass = hist.Hist.new.Reg(50, 0., 200.).Double()  # Massa invariante dos b-jatos
histo_bjets_inv_mass = hist.Hist.new.Reg(50, 0., 200.).Double()  # Massa invariante dos b-jatos

# Função para calcular a pseudorapidez (eta)
def get_eta(particle):
    p = np.sqrt(particle.momentum.x**2 + particle.momentum.y**2 + particle.momentum.z**2)
    if p == 0 or np.isclose(particle.momentum.z, p) or np.isclose(particle.momentum.z, -p):
        return np.nan
    return 0.5 * np.log((p + particle.momentum.z) / (p - particle.momentum.z))

# Função para calcular a separação Rjj entre dois jatos
def delta_r(jet, particle):
    delta_eta = jet.eta() - get_eta(particle)
    delta_phi = jet.phi() - np.arctan2(particle.momentum.y, particle.momentum.x)
    delta_phi = (delta_phi + np.pi) % (2 * np.pi) - np.pi  # Ajuste para o intervalo [-π, π]
    return np.sqrt(delta_eta**2 + delta_phi**2)

# Função para calcular a massa invariante de dois jatos
def invariant_mass(jet1, jet2):
    E_tot = jet1.e() + jet2.e()
    px_tot = jet1.px() + jet2.px()
    py_tot = jet1.py() + jet2.py()
    pz_tot = jet1.pz() + jet2.pz()
    return np.sqrt(E_tot**2 - (px_tot**2 + py_tot**2 + pz_tot**2))

# Função para calcular o deslocamento do vértice
def vertex_displacement(particle):
    vx = particle.production_vertex.position.x
    vy = particle.production_vertex.position.y
    vz = particle.production_vertex.position.z
    displacement = np.sqrt(vx**2 + vy**2 + vz**2)
    return displacement

# Função para calcular a significância do deslocamento (parâmetro de impacto)
def impact_parameter_significance(particle):
    displacement = vertex_displacement(particle)
    uncertainty = particle.production_vertex.position.t  # Supondo que t contém a incerteza
    return displacement / uncertainty if uncertainty > 0 else 0

# Função para calcular a massa do vértice secundário
def secondary_vertex_mass(jet, particles):
    tracks = [p for p in particles if delta_r(jet, p) < 0.4]  # Trilhas associadas ao jato
    px, py, pz, energy = 0, 0, 0, 0
    for t in tracks:
        px += t.momentum.x
        py += t.momentum.y
        pz += t.momentum.z
        energy += t.momentum.e
    mass = np.sqrt(max(energy**2 - (px**2 + py**2 + pz**2), 0))
    return mass

# Função para verificar se uma partícula é um múon ou elétron
def is_lepton(particle):
    return abs(particle.pid) in [11, 13]  # IDs de elétron (11) e múon (13)

# Função para identificar b-jatos
def identify_b_jets(jets, particles):
    b_jets = []
    for jet in jets:
        # Identificar múons e elétrons associados
        associated_leptons = [p for p in particles if is_lepton(p) and delta_r(jet, p) < 0.4]
        
        # Verificar critérios de deslocamento e massa do vértice
        if len(associated_leptons) > 0:
            mass = secondary_vertex_mass(jet, particles)
            if mass > 4.5:  # Critério aproximado para hadrons b
                b_jets.append(jet)
    return b_jets

# Função para analisar o evento e preencher os histogramas de b-jatos
def analyze_event(event):
    particles = [p for p in event.particles if p.status == 1]  # Partículas finais
    fj_particles = [fj.PseudoJet(p.momentum.x, p.momentum.y, p.momentum.z, p.momentum.e) for p in particles]
    
    jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)
    clusterer = fj.ClusterSequence(fj_particles, jet_def)
    jets = clusterer.inclusive_jets(10.0)  # Jatos com pT > 10 GeV
    
    # Identificar b-jatos
    b_jets = identify_b_jets(jets, particles)
    
    # Preencher histogramas de b-jatos
    for b_jet in b_jets:
        histo_bjets_pt.fill(b_jet.pt())  # Preencher o histograma de pT dos b-jatos
        histo_bjets_eta.fill(b_jet.eta())  # Preencher o histograma de eta dos b-jatos
        
        # Preencher histograma da massa invariante dos b-jatos
        if len(b_jets) >= 2:
            for i in range(len(b_jets)-1):
                for j in range(i+1, len(b_jets)):
                    mass = invariant_mass(b_jets[i], b_jets[j])
                    histo_bjets_inv_mass.fill(mass)

# Função para visualizar os histogramas de b-jatos
def plot_histograms():
    fig, ax = plt.subplots(1, 3, figsize=(18, 6))

    histo_bjets_pt.plot(ax=ax[0], ls="-", color="red", lw=3)
    ax[0].set(xlabel='pT dos b-Jatos (GeV)', ylabel='Contagem')

    histo_bjets_eta.plot(ax=ax[1], ls="-", color="blue", lw=3)
    ax[1].set(xlabel='Rapidez dos b-Jatos (Eta)', ylabel='Contagem')

    histo_bjets_inv_mass.plot(ax=ax[2], ls="-", color="green", lw=3)
    ax[2].set(xlabel='Massa Invariante dos b-Jatos (GeV)', ylabel='Contagem')

    plt.tight_layout()
    plt.show()

# Ler eventos do arquivo HEPMC
events = []
with hep.open(hepmc_file) as f:
    for event in f:
        events.append(event)

# Loop sobre os eventos e realizar a análise
for ievt, event in enumerate(events):
    if ievt % 1000 == 0:
        print(f"Processando evento {ievt} de {len(events)}...")
    analyze_event(event)

# Plotar os histogramas de b-jatos
plot_histograms()
```
