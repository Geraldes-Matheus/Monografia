```python
import pyhepmc as hep
import fastjet as fj
import numpy as np
import matplotlib.pyplot as plt

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Função para rastrear as partículas filhas do b até o estado final
def trace_b_decays(particle, traced_particles):
    if particle.status == 1:  # Se for uma partícula final
        traced_particles.append(particle)
    elif particle.status == 2:  # Se for uma partícula intermediária
        traced_particles.append(particle)  # Adiciona a partícula intermediária à lista
        for daughter in particle.children:
            trace_b_decays(daughter, traced_particles)  # Recursão nas filhas
    return traced_particles

# Função para ler o arquivo HEPMC, realizar o clustering de jatos e plotar a distância
def process_events_and_cluster_jets(hepmc_file, max_events=10):
    distances_b = []  # Lista para armazenar as distâncias médias dos jatos b ao vértice primário
    distances_non_b = []  # Lista para armazenar as distâncias médias dos jatos não b ao vértice primário

    with hep.open(hepmc_file) as f:
        # Itera sobre os eventos no arquivo
        for event_number, event in enumerate(f, 1):
            if event_number > max_events:
                break  # Só processa os 10 primeiros eventos

            # Verifica se o evento possui partículas
            if len(event.particles) > 0:
                particles = []
                b_particles = []  # Lista para armazenar as partículas de quark b encontradas
                jets_b = []  # Lista para armazenar jatos b identificados
                jets_non_b = []  # Lista para armazenar jatos não b identificados
                
                # Obter a posição do vértice primário (assumindo que seja o primeiro vértice)
                primary_vertex = event.vertices[0] if event.vertices else None
                if primary_vertex:
                    primary_vertex_position = np.array([primary_vertex.position.x,
                                                        primary_vertex.position.y,
                                                        primary_vertex.position.z])
                else:
                    print(f"Evento: {event_number} não possui vértice primário.")
                    continue

                # Processa as partículas do evento
                for p in event.particles:
                    if p.status == 1 or p.status == 2:
                        p_vector = fj.PseudoJet(p.momentum[0], p.momentum[1], p.momentum[2], p.momentum[3])
                        particles.append(p_vector)
                        
                        if abs(p.pid) == 5:  # Quark b (pid == 5 ou -5)
                            traced_b_particles = trace_b_decays(p, [])
                            b_particles.extend(traced_b_particles)  # Armazena as partículas de quark b e suas filhas
                        elif p.status == 2:
                            for parent in p.parents:
                                if abs(parent.pid) == 5:
                                    traced_b_particles = trace_b_decays(p, [])
                                    b_particles.extend(traced_b_particles)

                # Configuração de clustering de jatos
                jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)
                clusterer = fj.ClusterSequence(particles, jet_def)
                jets = clusterer.inclusive_jets(10.0)

                # Identificação e separação dos jatos b e não b
                for jet in jets:
                    jet_particles_positions = []
                    is_bjet = False  # Flag para verificar se é um jato b

                    for b_particle in b_particles:
                        # Calculando a distância delta_R entre o jato e a partícula filha do quark b
                        delta_R = jet.delta_R(fj.PseudoJet(b_particle.momentum[0], b_particle.momentum[1], b_particle.momentum[2], b_particle.momentum[3]))
                        if delta_R < 0.4:  # Usando delta_R para associar o jato à partícula filha do quark b
                            is_bjet = True  # Jato identificado como jato b
                            break  # Só conta o jato b uma vez

                    # Calculando a distância média das partículas dentro do jato ao vértice primário
                    for particle in jet.constituents():
                        particle_position = np.array([particle.px(), particle.py(), particle.pz()])
                        distance = np.linalg.norm(particle_position - primary_vertex_position)
                        jet_particles_positions.append(distance)

                    if jet_particles_positions:
                        mean_distance = np.mean(jet_particles_positions)
                        if is_bjet:
                            distances_b.append(mean_distance)  # Adiciona à lista de jatos b
                        else:
                            distances_non_b.append(mean_distance)  # Adiciona à lista de jatos não b

                # Imprime os resultados para o evento
                #print(f"Evento: {event_number}, Número de jatos no evento: {len(jets)}")

        # Plotar as distâncias médias dos jatos ao vértice primário
        plt.hist(distances_b, bins=50, histtype='step', linestyle='-', color='b', label='Jatos b')
        plt.hist(distances_non_b, bins=50, histtype='step', linestyle='-', color='r', label='Jatos não b')
        plt.xlabel('Distância média do Jato ao Vértice Primário [unidades de comprimento]')
        plt.ylabel('Número de Jatos')
        plt.title('Distribuição das Distâncias dos Jatos ao Vértice Primário')
        plt.legend()
        plt.show()

# Chama a função para processar os eventos e realizar o clustering de jets
process_events_and_cluster_jets(hepmc_file, max_events=1000)
```

```python
import pyhepmc as hep
import fastjet as fj
import numpy as np

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Função para rastrear as partículas filhas do b até o estado final
def trace_b_decays(particle, traced_particles):
    if particle.status == 1:  # Se for uma partícula final
        traced_particles.append(particle)
    elif particle.status == 2:  # Se for uma partícula intermediária
        traced_particles.append(particle)  # Adiciona a partícula intermediária à lista
        for daughter in particle.children:
            trace_b_decays(daughter, traced_particles)  # Recursão nas filhas
    return traced_particles

# Função para ler o arquivo HEPMC, realizar o clustering de jatos e rastrear partículas
def process_events_and_cluster_jets(hepmc_file, max_events=10):
    with hep.open(hepmc_file) as f:
        # Itera sobre os eventos no arquivo
        for event_number, event in enumerate(f, 1):
            if event_number > max_events:
                break  # Só processa os 10 primeiros eventos

            # Verifica se o evento possui partículas
            if len(event.particles) > 0:
                particles = []
                b_particles = []  # Lista para armazenar as partículas de quark b encontradas
                jets_b = []  # Lista para armazenar jatos b identificados
                
                # Obter a posição do vértice primário (assumindo que seja o primeiro vértice)
                primary_vertex = event.vertices[0] if event.vertices else None
                if primary_vertex:
                    primary_vertex_position = np.array([primary_vertex.position.x,
                                                        primary_vertex.position.y,
                                                        primary_vertex.position.z])
                else:
                    print(f"Evento: {event_number} não possui vértice primário.")
                    continue

                # Imprime o número do evento
                print(f"Evento {event_number}:")

                # Processa as partículas do evento
                for p in event.particles:
                    if p.status == 1 or p.status == 2:
                        p_vector = fj.PseudoJet(p.momentum[0], p.momentum[1], p.momentum[2], p.momentum[3])
                        particles.append(p_vector)
                        
                        if abs(p.pid) == 5:  # Quark b (pid == 5 ou -5)
                            traced_b_particles = trace_b_decays(p, [])
                            b_particles.extend(traced_b_particles)  # Armazena as partículas de quark b e suas filhas

                            # Aqui começamos a rastrear os jatos de b
                            # Jato de b começa com a partícula b e vai até a última partícula filha
                            jet_chain = [p.pid]  # Inicia o jato com a partícula b
                            for i in range(len(traced_b_particles) - 1):
                                current_particle = traced_b_particles[i]
                                next_particle = traced_b_particles[i + 1]
                                
                                # Calcula delta_R entre partículas consecutivas
                                delta_R = fj.PseudoJet(current_particle.momentum[0], current_particle.momentum[1], current_particle.momentum[2], current_particle.momentum[3]).delta_R(
                                    fj.PseudoJet(next_particle.momentum[0], next_particle.momentum[1], next_particle.momentum[2], next_particle.momentum[3])
                                )
                                
                                if delta_R < 0.4:  # Se delta_R < 0.4, as partículas estão associadas a um jato
                                    jet_chain.append(next_particle.pid)  # Adiciona a partícula ao jato
                                else:
                                    break  # Se não está associado, encerra o jato

                            # Agora imprimimos a cadeia do jato de b formada
                            print(f"  jato de b formado por: {' -> '.join(map(str, jet_chain))}")
                            
                        elif p.status == 2:
                            for parent in p.parents:
                                if abs(parent.pid) == 5:
                                    traced_b_particles = trace_b_decays(p, [])
                                    b_particles.extend(traced_b_particles)
                                    
                                    # Aqui começamos a rastrear os jatos de b para partículas intermediárias
                                    # Jato de b começa com a partícula b e vai até a última partícula filha
                                    jet_chain = [parent.pid]  # Inicia o jato com a partícula b
                                    for i in range(len(traced_b_particles) - 1):
                                        current_particle = traced_b_particles[i]
                                        next_particle = traced_b_particles[i + 1]
                                        
                                        # Calcula delta_R entre partículas consecutivas
                                        delta_R = fj.PseudoJet(current_particle.momentum[0], current_particle.momentum[1], current_particle.momentum[2], current_particle.momentum[3]).delta_R(
                                            fj.PseudoJet(next_particle.momentum[0], next_particle.momentum[1], next_particle.momentum[2], next_particle.momentum[3])
                                        )
                                        
                                        if delta_R < 0.4:  # Se delta_R < 0.4, as partículas estão associadas a um jato
                                            jet_chain.append(next_particle.pid)  # Adiciona a partícula ao jato
                                        else:
                                            break  # Se não está associado, encerra o jato

                                    # Agora imprimimos a cadeia do jato de b formada
                                    print(f"  jato de b formado por: {' -> '.join(map(str, jet_chain))}")

# Chama a função para processar os eventos e realizar o clustering de jets
process_events_and_cluster_jets(hepmc_file, max_events=10)
```

```python
import pyhepmc as hep
import fastjet as fj
import math

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Função para verificar se a partícula é filha de um quark b (verificando suas mães)
def is_secondary_vertex(particle):
    # Acesse as partículas mães através do atributo `parents`, que é uma lista
    parents = particle.parents
    for parent in parents:
        if abs(parent.pid) == 5:  # PDG do quark b
            return True
    return False

# Função para calcular a distância entre dois vértices (usando coordenadas 3D)
def calculate_distance(vertex1, vertex2):
    dx = vertex1[0] - vertex2[0]
    dy = vertex1[1] - vertex2[1]
    dz = vertex1[2] - vertex2[2]
    return math.sqrt(dx*dx + dy*dy + dz*dz)

# Função para processar os eventos e realizar o clustering de jatos
def process_events_and_cluster_jets(hepmc_file):
    with hep.open(hepmc_file) as f:
        # Itera sobre os eventos no arquivo
        for event in f:
            # Verifica se o evento possui partículas
            if len(event.particles) > 0:
                # Criar uma lista de partículas para o FastJet
                particles = []
                b_particles = []  # Lista para armazenar as partículas do b quark
                for p in event.particles:
                    # Só considera partículas finais
                    if p.status == 1:
                        # FastJet usa 4-vetores: (px, py, pz, E)
                        p_vector = fj.PseudoJet(p.momentum[0], p.momentum[1], p.momentum[2], p.momentum[3])
                        particles.append(p_vector)
                        
                        # Se a partícula for filha de um quark b, adiciona na lista de b_particles
                        if is_secondary_vertex(p):
                            b_particles.append(p)
                
                # Configurar o algoritmo de jet clustering (por exemplo, kT com R=0.4)
                jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)
                clusterer = fj.ClusterSequence(particles, jet_def)
                
                # Obter os jets formados
                jets = clusterer.inclusive_jets(10.0)  # Corte de pT mínimo dos jets
                
                print(f"\nJets encontrados no evento:")
                for jet in jets:
                    print(f"Jet: pT = {jet.pt()}, eta = {jet.eta()}, phi = {jet.phi()}")
                    
                    # Verifica se o jet contém partículas de b
                    is_b_jet = False
                    for b_particle in b_particles:
                        # Verifica se a partícula está dentro do jet
                        if jet.delta_R(fj.PseudoJet(b_particle.momentum[0], b_particle.momentum[1], b_particle.momentum[2], b_particle.momentum[3])) < 0.4:
                            is_b_jet = True
                            break
                    
                    if is_b_jet:
                        print(f"Jet é um jato de b.")
                    
                    # Para cada jet, verificamos se ele está no vértice secundário
                    for b_particle in b_particles:
                        # Aqui você pode calcular a distância entre o vértice primário e o vértice secundário
                        primary_vertex = (event.vertices[0].position[0], event.vertices[0].position[1], event.vertices[0].position[2])
                        
                        # Acessa o atributo `production_vertex` diretamente
                        secondary_vertex = (b_particle.production_vertex.position[0], b_particle.production_vertex.position[1], b_particle.production_vertex.position[2])
                        
                        # Distância entre o vértice primário e o secundário
                        distance = calculate_distance(primary_vertex, secondary_vertex)
                        print(f"Distância entre vértices primário e secundário: {distance} unidades.")
            else:
                print("O evento não possui partículas.")
            
            break  # Apenas processa o primeiro evento

# Chama a função para processar o evento e realizar o clustering de jets
process_events_and_cluster_jets(hepmc_file)
```
