```python
import pyhepmc as hep

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Função para ler o arquivo HEPMC e mostrar os atributos dos vértices, end_vertex e production_vertex
def show_vertex_info(hepmc_file):
    with hep.open(hepmc_file) as f:
        # Itera sobre os eventos no arquivo
        for event in f:
            
            # Verifica se o evento possui partículas e exibe as informações dos vértices das partículas
            if len(event.particles) > 0:
                for particle in event.particles:
                    print("\nPartícula:")
                    print("PID:", particle.pid)  # Identificador da partícula
                    print("Status:", particle.status)  # Status da partícula (final ou intermediária)
                    
                    # Exibe o vértice de produção da partícula
                    if particle.production_vertex:
                        print("Vértice de produção da partícula:")
                        print("Posição do vértice de produção (x, y, z, t):", particle.production_vertex.position)
                    else:
                        print("A partícula não tem vértice de produção.")
                    
                    # Exibe o vértice final (end_vertex) da partícula
                    if particle.end_vertex:
                        print("Vértice final da partícula:")
                        print("Posição do vértice final (x, y, z, t):", particle.end_vertex.position)
                    else:
                        print("A partícula não tem vértice final.")
            else:
                print("O evento não possui partículas.")
            
            break  # Apenas processa o primeiro evento
# Chama a função para exibir as informações dos vértices
show_vertex_info(hepmc_file)
```

```python
import pyhepmc as hep
import fastjet as fj

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Função para ler o arquivo HEPMC e realizar o clustering de jatos
def process_events_and_cluster_jets(hepmc_file):
    with hep.open(hepmc_file) as f:
        # Itera sobre os eventos no arquivo
        for event in f:
            # Verifica se o evento possui partículas
            if len(event.particles) > 0:
                # Criar uma lista de partículas para o FastJet
                particles = []
                for p in event.particles:
                    # Só considera partículas que são partículas de interesse para jet clustering
                    if p.status == 1:  # Partículas finais
                        # FastJet usa 4-vetores: (px, py, pz, E)
                        p_vector = fj.PseudoJet(p.momentum[0], p.momentum[1], p.momentum[2], p.momentum[3])
                        particles.append(p_vector)
                
                # Configurar o algoritmo de jet clustering (por exemplo, kT com R=0.6)
                jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.4)
                clusterer = fj.ClusterSequence(particles, jet_def)
                
                # Obter os jets formados
                jets = clusterer.inclusive_jets(10.0)  # Corte de pT mínimo dos jets
                
                print(f"\nJets encontrados no evento:")
                for jet in jets:
                    print(f"Jet: pT = {jet.pt()}, eta = {jet.eta()}, phi = {jet.phi()}")
            else:
                print("O evento não possui partículas.")
            
            break  # Apenas processa o primeiro evento

# Chama a função para processar o evento e realizar o clustering de jets
process_events_and_cluster_jets(hepmc_file)
```

```python
import pyhepmc as hep
import fastjet as fj

# Caminho para o arquivo .hepmc
hepmc_file = 'tag_1_pythia8_events.hepmc'

# Função para ler o arquivo HEPMC e realizar o clustering de jatos
def process_events_and_cluster_jets(hepmc_file):
    with hep.open(hepmc_file) as f:
        # Itera sobre os eventos no arquivo
        for event in f:
            # Verifica se o evento possui partículas
            if len(event.particles) > 0:
                # Criar uma lista de partículas para o FastJet
                particles = []
                for p in event.particles:
                    # Só considera partículas que são partículas de interesse para jet clustering
                    if p.status == 1:  # Partículas finais
                        # FastJet usa 4-vetores: (px, py, pz, E)
                        p_vector = fj.PseudoJet(p.momentum[0], p.momentum[1], p.momentum[2], p.momentum[3])
                        particles.append(p_vector)
                
                # Configurar o algoritmo de jet clustering (por exemplo, kT com R=0.6)
                jet_def = fj.JetDefinition(fj.antikt_algorithm, 0.6)
                clusterer = fj.ClusterSequence(particles, jet_def)
                
                # Obter os jets formados
                jets = clusterer.inclusive_jets(20.0)  # Corte de pT mínimo dos jets
                
                # Exibir métodos e atributos uma única vez
                methods = set()  # Usando set para evitar repetições
                attributes = set()

                for jet in jets:
                    # Listar métodos e atributos separadamente
                    for attr in dir(jet):
                        if callable(getattr(jet, attr)):
                            methods.add(attr)
                        else:
                            attributes.add(attr)
                
                print(f"\nAtributos e Métodos do Jet (PseudoJet):")
                
                # Exibir Métodos
                print("\nMétodos:")
                for method in sorted(methods):
                    print(f"  {method}")

                # Exibir Atributos
                print("\nAtributos:")
                for attribute in sorted(attributes):
                    try:
                        value = getattr(jet, attribute)
                        print(f"  {attribute} = {value}")
                    except Exception as e:
                        print(f"  {attribute} = Erro ao acessar: {e}")
            
            else:
                print("O evento não possui partículas.")
            
            break  # Apenas processa o primeiro evento

# Chama a função para processar o evento e realizar o clustering de jets
process_events_and_cluster_jets(hepmc_file)
```
