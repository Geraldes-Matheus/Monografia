```python
# Abrindo todos os arquivos e acessando a tree 'Events'
data = [uproot.open(arquivo) for arquivo in arquivos]
trees = [d['Events;1'] for d in data]

# Inicializando listas para armazenar as variáveis de todos os arquivos
muon_pts = []
muon_etas = []
muon_phis = []
muon_masses = []
nmuons = []
muon_charges = []

# Extraindo as variáveis de cada arquivo e concatenando
for tree in trees:
    muon_pts.extend(tree['Muon_pt'].array())  # Usando extend() para adicionar à lista
    muon_etas.extend(tree['Muon_eta'].array())  # Usando extend() para adicionar à lista
    muon_phis.extend(tree['Muon_phi'].array())  # Usando extend() para adicionar à lista
    muon_masses.extend(tree['Muon_mass'].array())  # Usando extend() para adicionar à lista
    nmuons.extend(tree['nMuon'].array())  # Usando extend() para adicionar à lista
    muon_charges.extend(tree['Muon_charge'].array())  # Usando extend() para adicionar à lista
```

```python
import awkward as ak

# Caso nmuons seja uma lista, converta-a para uma array Awkward
if isinstance(nmuons, list):
    nmuons = ak.Array(nmuons)

# Verifique o tipo de nmuons após a conversão
print(f"Tipo de nmuons após conversão: {type(nmuons)}")

# Verifique o número de eventos antes de qualquer corte
num_events_initial = len(nmuons)
print(f"Número inicial de eventos: {num_events_initial}")

# Seleção do número de múons (número de múons = 4)
cut_nmuons = (nmuons == 4)

# Verifique se cut_nmuons é uma máscara válida (deve ser um array Awkward)
print(f"Tipo de cut_nmuons: {type(cut_nmuons)}")
print(f"Exemplo de cut_nmuons: {ak.to_list(cut_nmuons)[:5]}")  # Exibindo os primeiros 5 valores

# Soma os eventos após o corte de número de múons (= 4)
num_events_after_nmuons = ak.sum(cut_nmuons)
print(f"Número de eventos após o corte de número de múons (= 4): {num_events_after_nmuons}")

# Seleção de múons de cargas opostas
# Usando ak.combinations para verificar múons de cargas opostas
muon_pairs = ak.combinations(muon_charges, 2, axis=1)  # Cria pares de múons
cut_osmuons = (muon_pairs['0'] != muon_pairs['1'])  # Verifica se as cargas são opostas

# Verifique se cut_osmuons é uma máscara válida
print(f"Tipo de cut_osmuons: {type(cut_osmuons)}")
print(f"Exemplo de cut_osmuons: {ak.to_list(cut_osmuons)[:5]}")  # Exibindo os primeiros 5 valores

# Verifique se pelo menos um par de múons com cargas opostas está presente por evento
cut_osmuons_any = ak.any(cut_osmuons, axis=1)

# Soma os eventos após o corte de múons com cargas opostas
num_events_after_osmuons = ak.sum(cut_nmuons & cut_osmuons_any)
print(f"Número de eventos após o corte de cargas opostas: {num_events_after_osmuons}")

# Combinação de todas as máscaras
cut_combined = cut_nmuons & cut_osmuons_any
num_events_after_combined = ak.sum(cut_combined)
print(f"Número de eventos após todos os cortes (combinação completa): {num_events_after_combined}")
```

```python
import awkward as ak

# Converter as listas para Awkward arrays (se ainda não foram convertidas)
muon_pts = ak.Array(muon_pts)
muon_etas = ak.Array(muon_etas)
muon_phis = ak.Array(muon_phis)
muon_masses = ak.Array(muon_masses)
nmuons = ak.Array(nmuons)

# Agora, podemos aplicar os cortes
muon_pt_sel = muon_pts[cut_combined]
muon_eta_sel = muon_etas[cut_combined]
muon_phi_sel = muon_phis[cut_combined]
muon_mass_sel = muon_masses[cut_combined]
nmu_sel = nmuons[cut_combined]

# Verifique os arrays selecionados (pode imprimir os primeiros 5 elementos para inspeção)
print("muon_pt_sel:", muon_pt_sel[:5])
print("muon_eta_sel:", muon_eta_sel[:5])
print("muon_phi_sel:", muon_phi_sel[:5])
print("muon_mass_sel:", muon_mass_sel[:5])
print("nmu_sel:", nmu_sel[:5])
```

## Criando os 4-Vetores

vector.array: Cria um array de 4-vetores com os componentes pt, eta, phi e mass dos múons. Vamos criar os 4-vetores dos múons utilizando as componentes cinemáticas que você já aplicou os cortes em função de (pt, eta, phi, mass).

*    Criar o 4-vetor para cada múon. .
*    Somar os 4-vetores: Após selecionar os dois múons, somamos os seus 4-vetores para obter o 4-vetores do sistema


```python
#  Criar os quatro-vetores dos múons
muon1_4vecs = vector.zip({
    "pt": muon1_pt_cut,
    "eta": muon1_eta_cut,
    "phi": muon1_phi_cut,
    "mass": muon1_mass_cut
})

muon2_4vecs = vector.zip({
    "pt": muon2_pt_cut,
    "eta": muon2_eta_cut,
    "phi": muon2_phi_cut,
    "mass": muon2_mass_cut
})
```

```python
dimuon1 = muon1_4vecs + muon2_4vecs
print(dimuon1.fields)
```

```python
# Extraindo as propriedades do quadrivetor do par
pt_dimu1 = dimuon1.pt
phi_dimu1 = dimuon1.phi
eta_dimu1 = dimuon1.eta
mass_dimu1 = dimuon1.mass
```

```python
import awkward as ak

# Função para verificar se o muão ou antimuão veio de um Z
def identify_muons_from_Z(tree):
    muons_from_Z = 0  # Contador de muões/antimuões que vieram de um Z

    # Acessando a lista de partículas
    nGenPart = tree["nGenPart"].array()
    GenPart_pdgId = tree["GenPart_pdgId"].array()
    GenPart_genPartIdxMother = tree["GenPart_genPartIdxMother"].array()
    
    # Loop sobre todas as partículas
    for i in range(len(nGenPart)):  # Usando len(nGenPart) ao invés de shape
        pdgId = GenPart_pdgId[i]
        genPartIdxMother = GenPart_genPartIdxMother[i]

        # Verifica se a partícula é um muão ou antimuão
        if ak.any(pdgId == 13) or ak.any(pdgId == -13):  # Usando ak.any() para comparação
            # Verifica se a partícula mãe é um Z (pdgId = 23)
            if ak.any(GenPart_pdgId[genPartIdxMother] == 23):
                muons_from_Z += 1  # Incrementa o contador se a partícula veio de um Z

    return muons_from_Z

# Exemplo de chamada da função
muons_from_Z = identify_muons_from_Z(tree)

print(f"Total de muões/antimuões provenientes de um Z: {muons_from_Z}")
```
